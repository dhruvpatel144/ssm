```python
def forward(pi, A, B, outcome):
    length = len(outcome)
    number_of_states = len(A)
    alpha = [[0] * number_of_states for i in range(length)]
    for i in range(number_of_states):
        alpha[0][i] = pi[i] * B[i][outcome[0]]
        # alpha = alpha.at(0,i).set(pi[i]*B[i,outcome[0]])

    for i in range(1, length):
        for j in range(number_of_states):
            sum = 0
            for q in range(number_of_states):
                sum += alpha[i - 1][q] * A[q][j]
            # alpha = alpha.at(i,j).set(sum*B[j,outcome[i]])
            alpha[i][j] = sum * B[j][outcome[i]]

    return alpha


def backward(pi, A, B, outcome):
    length = len(outcome)
    number_of_states = len(A)
    beta = [[0] * number_of_states for i in range(length)]
    for st in range(number_of_states):
        beta[length - 1][st] = 1

    for t in range(length - 2, -1, -1):
        for st in range(number_of_states):
            sum = 0
            for j in range(number_of_states):
                sum += beta[t + 1][j] * A[st][j] * B[j][outcome[t + 1]]
            beta[t][st] = sum

    return beta


def expectation(pi, A, B, outcome):
    length = len(outcome)
    number_of_states = len(A)
    alpha = forward(pi, A, B, outcome)
    beta = backward(pi, A, B, outcome)
    exp = []
    for t in range(length - 1):
        e = [[0] * number_of_states for i in range(number_of_states)]
        sum = 0
        for i in range(number_of_states):
            for j in range(number_of_states):
                e[i][j] = alpha[t][i] * A[i][j] * B[j][outcome[t + 1]] * beta[t + 1][j]
                sum += e[i][j]
        for i in range(number_of_states):
            for j in range(number_of_states):
                if sum != 0:
                    e[i][j] = e[i][j] / sum
        exp.append(e)

    gamma = [[0] * number_of_states for i in range(length - 1)]
    for t in range(length - 1):
        for i in range(number_of_states):
            sum = 0
            for j in range(number_of_states):
                sum += exp[t][i][j]
            gamma[t][i] = sum
    return exp, gamma


def maximization(pi, A, B, outcomes):
    n = len(outcomes)
    length = len(outcomes[0])
    p = len(B[0])
    number_of_states = len(A)
    exps = []
    gammas = []
    for i in range(n):
        ex, gm = expectation(pi, A, B, outcomes[i])
        exps.append(ex)
        gammas.append(gm)

    pi_updated = []
    for i in range(number_of_states):
        sum = 0
        for j in range(n):
            sum += gammas[j][0][i]
        pi_updated.append(sum / n)

    A_updated = [[0] * number_of_states for i in range(number_of_states)]
    for j in range(number_of_states):
        v = 0
        for m in range(number_of_states):
            sum = 0
            for i in range(n):
                for t in range(length - 1):
                    sum += exps[i][t][j][m]
                    v += exps[i][t][j][m]
            A_updated[j][m] = sum

        for m in range(number_of_states):
            if v != 0:
                A_updated[j][m] = A_updated[j][m] / v

    B_updated = [[0] * p for i in range(number_of_states)]
    for j in range(number_of_states):
        for m in range(p):
            sum = 0
            v = 0
            for i in range(n):
                for t in range(length - 1):
                    if outcomes[i][t] == m:
                        sum += gammas[i][t][j]
                    v += gammas[i][t][j]
            if v != 0:
                B_updated[j][m] = sum / v

    return pi_updated, A_updated, B_updated
```
